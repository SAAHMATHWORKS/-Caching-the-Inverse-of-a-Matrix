```
X_original = pd.read_csv('train.csv', index_col='Id')
X_test_full = pd.read_csv('test.csv', index_col='Id')

X = X_original.copy()
```

```
X.head()
```

```
#Create single wilderness area and soil_type field (reverse the one hot encoding)
Wilderness_Area_Fields = ['Wilderness_Area'+ str(i) for i in range(1,5)]
Soil_Type = [col for col in X if col.startswith('Soil')]
X['Wilderness_Area'] = X[Wilderness_Area_Fields].idxmax(1)
X_test_full['Wilderness_Area'] = X_test_full[Wilderness_Area_Fields].idxmax(1)
X.drop(Wilderness_Area_Fields, axis = 1, inplace = True)
X_test_full.drop(Wilderness_Area_Fields, axis = 1, inplace = True)
X['Soil_Type'] = X[Soil_Type].idxmax(1)
X_test_full['Soil_Type'] = X_test_full[Soil_Type].idxmax(1)
X.drop(Soil_Type, axis = 1, inplace = True)
X_test_full.drop(Soil_Type, axis = 1, inplace = True)
X.head()
```

```
X_test_full.head()
```

```
X['EV_DTH'] = (X.Elevation - X.Vertical_Distance_To_Hydrology)
X_test_full['EV_DTH'] = (X_test_full.Elevation - X_test_full.Vertical_Distance_To_Hydrology)


X['EH_DTH'] = (X.Elevation -  (X.Horizontal_Distance_To_Hydrology *0.2))
X_test_full['EH_DTH'] = (X_test_full.Elevation -  (X_test_full.Horizontal_Distance_To_Hydrology *0.2))

X['Dis_To_Hy'] = (((X.Horizontal_Distance_To_Hydrology **2) + (X.Vertical_Distance_To_Hydrology **2))**0.5)
X_test_full['Dis_To_Hy'] = (((X_test_full.Horizontal_Distance_To_Hydrology **2) + (X_test_full.Vertical_Distance_To_Hydrology **2))**0.5)

X['HyF_1'] = (X.Horizontal_Distance_To_Hydrology + X.Horizontal_Distance_To_Fire_Points)
X_test_full['HyF_1'] = (X_test_full.Horizontal_Distance_To_Hydrology + X_test_full.Horizontal_Distance_To_Fire_Points)

X['HyF_2'] = (X.Horizontal_Distance_To_Hydrology - X.Horizontal_Distance_To_Fire_Points)
X_test_full['HyF_2'] = (X_test_full.Horizontal_Distance_To_Hydrology - X_test_full.Horizontal_Distance_To_Fire_Points)

X['HyR_1'] = (X.Horizontal_Distance_To_Hydrology + X.Horizontal_Distance_To_Roadways)
X_test_full['HyR_1'] = (X_test_full.Horizontal_Distance_To_Hydrology + X_test_full.Horizontal_Distance_To_Roadways)

X['HyR_2'] = (X.Horizontal_Distance_To_Hydrology - X.Horizontal_Distance_To_Roadways)
X_test_full['HyR_2'] = (X_test_full.Horizontal_Distance_To_Hydrology - X_test_full.Horizontal_Distance_To_Roadways)


X['FiR_1'] = (X.Horizontal_Distance_To_Fire_Points + X.Horizontal_Distance_To_Roadways)
X_test_full['FiR_1'] = (X_test_full.Horizontal_Distance_To_Fire_Points + X_test_full.Horizontal_Distance_To_Roadways)

X['FiR_1'] = (X.Horizontal_Distance_To_Fire_Points - X.Horizontal_Distance_To_Roadways)
X_test_full['FiR_1'] = (X_test_full.Horizontal_Distance_To_Fire_Points - X_test_full.Horizontal_Distance_To_Roadways)

X['Avg_shade'] = ((X.Hillshade_9am + X.Hillshade_Noon + X.Hillshade_3pm) /3)
X_test_full['Avg_shade'] = ((X_test_full.Hillshade_9am + X_test_full.Hillshade_Noon + X_test_full.Hillshade_3pm) /3)

X['Morn_noon_int'] = ((X.Hillshade_9am + X.Hillshade_Noon) / 2)
X_test_full['Morn_noon_int'] = ((X_test_full.Hillshade_9am + X_test_full.Hillshade_Noon) / 2)

X['noon_eve_int'] = ((X.Hillshade_3pm + X.Hillshade_Noon) / 2)
X_test_full['noon_eve_int'] = ((X_test_full.Hillshade_3pm + X_test_full.Hillshade_Noon) / 2)

X['Slope2'] = np.sqrt(X.Horizontal_Distance_To_Hydrology**2 + X.Vertical_Distance_To_Hydrology**2)
X_test_full['Slope2'] = np.sqrt(X_test_full.Horizontal_Distance_To_Hydrology**2 + X_test_full.Vertical_Distance_To_Hydrology**2)
```

```
X['Cover_Type'] = X['Cover_Type'] - 1
```

```
# Split the training data
X_train, X_val, y_train, y_val = train_test_split(X.drop('Cover_Type', axis = 1), X['Cover_Type'], 
                                                  test_size = 0.50, shuffle = True)
```

```
X_train.head()
```

```
embed_cols=['Soil_Type', 'Wilderness_Area']
num_col = [c for c in X_train.columns if (not c in embed_cols)]
num_col
```

```
from sklearn.preprocessing import RobustScaler
#scale = RobustScaler(quantile_range = (0.1,0.9))
scale = StandardScaler()
scale.fit(X_train[num_col])
X_train[num_col] = scale.transform(X_train[num_col])
X_val[num_col] = scale.transform(X_val[num_col])
X_test_full[num_col] = scale.transform(X_test_full[num_col] )
```

```
X_train.head()
```

```
X_train.shape
```

```
X_train, X_val = X_train.align(X_val, join='left', axis=1)
X_train, X_test_full = X_train.align(X_test_full, join='left', axis=1)
```

```
embed_cols=['Soil_Type', 'Wilderness_Area']

#converting data to list format to match the network structure
def preproc(X_train, X_val, X_test_full):

    input_list_train = []
    input_list_val = []
    input_list_test = []
    #the cols to be embedded: rescaling to range [0, # values)
    for c in embed_cols:
        raw_vals = np.unique(X_train[c])
        val_map = {}
        for i in range(len(raw_vals)):
            val_map[raw_vals[i]] = i       
        input_list_train.append(X_train[c].map(val_map).values)
        input_list_val.append(X_val[c].map(val_map).fillna(8).values)
        input_list_test.append(X_test_full[c].map(val_map).fillna(8).values)
     
    #the rest of the columns
    other_cols = [c for c in X_train.columns if (not c in embed_cols)]
    input_list_train.append(X_train[other_cols].values)
    input_list_val.append(X_val[other_cols].values)
    input_list_test.append(X_test_full[other_cols].values)
    
    return input_list_train, input_list_val, input_list_test
```

```
input_list_train, input_list_val, input_list_test= preproc(X_train, X_val, X_test_full)
```

```
for categorical_var in embed_cols:
    
    cat_emb_name= categorical_var.replace(" ", "")+'_Embedding'
  
    no_of_unique_cat  = X_train[categorical_var].nunique()
    embedding_size = int(min(np.ceil((no_of_unique_cat)/2), 50 ))
    
    print(categorical_var ,X[categorical_var].nunique(), 'unique values presents in the train data')
    print('Categorical Variable:', categorical_var,
        'Unique Categories:', no_of_unique_cat,
        'Embedding Size:', embedding_size)
```

```
from keras.optimizers import SGD
from sklearn.preprocessing import LabelEncoder
from keras.callbacks import LearningRateScheduler
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, RMSprop, Adadelta, Adamax
from sklearn.preprocessing import LabelEncoder
from keras.callbacks import LearningRateScheduler
 
# learning rate schedule
def step_decay(epoch):
	initial_lrate = 0.1
	drop = 0.5
	epochs_drop = 10.0
	lrate = initial_lrate * 1e-8 * 10**(epoch / 20)
	return lrate
```

```
from keras.layers import BatchNormalization
input_models=[]
output_embeddings=[]
for categorical_var in embed_cols:
    
    #Name of the categorical variable that will be used in the Keras Embedding layer
    cat_emb_name= categorical_var.replace(" ", "")+'_Embedding'
  
    # Define the embedding_size
    no_of_unique_cat  = X_train[categorical_var].nunique()
    embedding_size = int(min(np.ceil((no_of_unique_cat)/2), 50 ))
  
    #One Embedding Layer for each categorical variable
    input_model = Input(shape=(1,))
    output_model = Embedding(no_of_unique_cat, embedding_size, name=cat_emb_name)(input_model)
    output_model = Reshape(target_shape=(embedding_size,))(output_model)    
  
    #Appending all the categorical inputs
    input_models.append(input_model)
  
    #Appending all the embeddings
    output_embeddings.append(output_model)
  
#Other non-categorical data columns (numerical). 
#I define single another network for the other columns and add them to our models list.
input_numeric = Input(shape=(len(num_col),))
embedding_numeric = Dense(50)(input_numeric) 
input_models.append(input_numeric)
output_embeddings.append(embedding_numeric)

#At the end we concatenate altogther and add other Dense layers
output = Concatenate()(output_embeddings)
output = Dense(248)(output)
output = BatchNormalization(momentum=0.0)(output)
output = Activation('relu')(output)
output= Dropout(0.4)(output)
output = Dense(124)(output)
output = BatchNormalization(momentum=0.0)(output)
output = Activation('relu')(output)
output= Dropout(0.3)(output)
output = Dense(7, activation='softmax')(output)
#, kernel_initializer="uniform"
model = Model(inputs=input_models, outputs=output)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
X_train_list,X_val_list,X_test_list = preproc(X_train,X_val, X_test_full)
```

```
history  =  model.fit(X_train_list,y_train,validation_data=(X_val_list,y_val) , epochs =  1000, batch_size=500,  verbose= 2)
```

```
preds_val = model.predict(X_val_list)
preds_val = np.argmax(preds_val, axis = 1)
preds_val
```
